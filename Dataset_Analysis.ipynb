```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load dataset
df = pd.read_csv('FindTaskCorpus.csv')

# Define a function to plot bar and pie charts
def plot_distribution(unique_values, counts, title):
    # Bar Chart
    plt.figure(figsize=(10, 5))
    plt.bar(unique_values, counts, color='skyblue', alpha=0.7)
    plt.xlabel('Categories')
    plt.ylabel('Counts')
    plt.title(f'Bar Chart of {title}')
    plt.xticks(unique_values)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

    # Pie Chart
    plt.figure(figsize=(7, 7))
    plt.pie(counts, labels=unique_values, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
    plt.title(f'Pie Chart of {title}')
    plt.axis('equal')  # Equal aspect ratio ensures the pie is drawn as a circle
    plt.show()

# Function to handle NaN values and compute distributions
def process_and_plot(column, title):
    # Replace NaN with a specific label for clarity, or drop NaN values
    column = column.fillna('NaN')  # Replace NaN with a string 'NaN'
    unique_values, counts = np.unique(column, return_counts=True)
    print(f"{title} Distribution:", dict(zip(unique_values, counts)))
    plot_distribution(unique_values, counts, title)

# eldOtHelBelief distribution
process_and_plot(df['eldOtHelBelief (0:indeterminate, 1:OK, 2:mismatch, 3:needs disambiguation)'], "eldOtHelBelief")

# eldLHelBelief distribution
process_and_plot(df['eldLHelBelief (0:indeterminate, 1:OK, 2:mismatch, 3:needs disambiguation)'], "eldLHelBelief")

# eldOHelBelief distribution
process_and_plot(df['eldOHelBelief (0:indeterminate, 1:OK, 2:mismatch, 3:needs disambiguation)'], "eldOHelBelief")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('FindTaskCorpus.csv')

# Define a function to plot bar and pie charts using pandas
def plot_distribution(column, title):
    # Count frequencies of each class
    counts = column.value_counts(dropna=False)  # Include NaN in the count

    # Bar chart
    plt.figure(figsize=(10, 6))
    counts.plot(kind='bar', color='skyblue', alpha=0.7)
    plt.title(f'Frequency of {title}')
    plt.xlabel('Category')
    plt.ylabel('Frequency')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    # Pie chart
    plt.figure(figsize=(8, 8))
    counts.plot(kind='pie', autopct='%1.1f%%', colors=plt.cm.Paired.colors)
    plt.title(f'Proportion of {title}')
    plt.ylabel('')  # Remove y-label for aesthetics
    plt.tight_layout()
    plt.show()

# Plot for eldOtHelBelief
plot_distribution(df['eldOtHelBelief (0:indeterminate, 1:OK, 2:mismatch, 3:needs disambiguation)'], "eldOtHelBelief")

# Plot for eldLHelBelief
plot_distribution(df['eldLHelBelief (0:indeterminate, 1:OK, 2:mismatch, 3:needs disambiguation)'], "eldLHelBelief")

# Plot for eldOHelBelief
plot_distribution(df['eldOHelBelief (0:indeterminate, 1:OK, 2:mismatch, 3:needs disambiguation)'], "eldOHelBelief")
```
